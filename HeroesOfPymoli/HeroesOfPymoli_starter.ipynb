```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Count the total number of players
player = len(purchase_data["SN"].value_counts())
total_players = pd.DataFrame([player], columns = ["Total Players"])
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#number of unique items
unique_item = len(purchase_data["Item Name"].unique())

#averge purchase price
avg_purchase_price = purchase_data["Price"].mean()

#total number of purchase
total_purchase = len(purchase_data["Item Name"])

#total revenue
Revenue = purchase_data["Price"].sum()

# Create new DataFrame
Purchasing_Analysis = pd.DataFrame({"Number of Unique Items": [unique_item],
                                           "Average Purchase Price": [avg_purchase_price],
                                           "Total Number of Purchases": [total_purchase],
                                           "Total Revenue": [Revenue]})
Purchasing_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# GroupBy gender
gender = purchase_data[["SN", "Gender"]]
gender = gender.drop_duplicates()

counts = gender["Gender"].value_counts()

# Total Count and Percentage
total_counts = [counts[0], counts[1], counts[2]]
percentage = [round((counts[0]/player)*100,2),round((counts[1]/player)*100,2),round((counts[2]/player)*100,2)]

gender_demographics = pd.DataFrame({"Number of Players": total_counts, "Percentage of Players": percentage})
gender_demographics.index = (["Male", "Female", "Other / Non-Disclosed"])
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# GroupBy Gender
gender = purchase_data.groupby(["Gender"])

# purchase count
purchase_count = gender["SN"].count()

#avg purchase price
avg_price = gender["Price"].mean()

#avg puchase total
total_avg_price = gender["Price"].sum()


# Create new DataFrame
Purchasing_Analysis_Gender = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price": avg_price,
                              "Total Purchase Value": total_avg_price})
Purchasing_Analysis_Gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Bins for ages
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 99]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#database for age
purchase_data["Age Range"]= pd.cut(purchase_data["Age"], age_bins, labels=group_names)

#Total count and percentage
age_total = purchase_data["Age Range"].value_counts()
age_percentage = round(age_total / player * 100,2)
age_demographics = pd.DataFrame({"Number of Players": age_total, "Percentage of Players": age_percentage})
age_demographics

age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin the Purchasing Data
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

# Run basic calculations
age_purchase_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"]
age_average = round(purchase_data.groupby(["Age Ranges"]).mean()["Price"],2)
age_counts = purchase_data.groupby(["Age Ranges"]).count()["Price"]


# Convert to DataFrame
age_data = pd.DataFrame({"Purchase Count": age_counts, "Average Purchase Price": age_average, "Total Purchase Value": age_purchase_total})
age_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#find top spender 
top_spender = purchase_data["Item ID"].groupby(purchase_data['SN']).count()

# create a dataframe 
top_spender= pd.DataFrame(data=top_spender)
top_spender.columns = ["Purchase Count"]

top_spender["Average Purchase Price"] = purchase_data["Price"].groupby(purchase_data["SN"]).mean()
top_spender["Total Purchase Value"] = purchase_data["Price"].groupby(purchase_data["SN"]).sum()

#sort in descending order
top_spender.sort_values(by=["Total Purchase Value"], ascending=False, inplace=True)
top_spender.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#find most popular item 
popular_item = purchase_data.groupby(["Item ID", "Item Name","Price"])["Price"].agg(["count","sum"])
popular_item.columns = ["Purchase Count", "Total Purchase Value"]

# set the index back at zero 
popular_item.reset_index(inplace=True)
popular_item.set_index(["Item ID","Item Name"] ,inplace=True)

popular_item = popular_item[["Purchase Count", "Price", "Total Purchase Value"]]

#Sort in descending order
most_popular_item = popular_item.sort_values(by="Purchase Count", ascending=False)
most_popular_item.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#sort in descending order 
most_profitable_items = popular_item.sort_values(by='Total Purchase Value', ascending=False)
most_profitable_items.head()
```
